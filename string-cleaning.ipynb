## string.split()

```python
def extract_last_word(element):
    element = str(element)
    splitted = element.split()
    return splitted[-1]

merged['Currency Apply'] = merged["CurrencyUnit"].apply(extract_last_word)

print(merged['Currency Apply'].head())
```

## regex (regular expressions)

 A regular expression is a sequence of characters that describes a search pattern, used to match characters in a string:
 ![ ](regex.png)

```python
#The brackets, [], indicate that either "national accounts" or "National accounts" should produce a match.
pattern = r"[Nn]ational accounts"
national_accounts = merged['SpecialNotes'].str.contains(pattern, regex = True)
print(national_accounts.head(10))
#I the Series.str.contains() method to see if a specific phrase appeared in a series.
# The result was a series containing True, False, and missing values.
```

```python
pattern = r"[Nn]ational accounts"
national_accounts = merged['SpecialNotes'].str.contains(pattern, na=False,regex=True)
#the na=False parameter makes the NaN values be read as false.

merged_national_accounts = merged[national_accounts]
print(merged_national_accounts.head())
```

```python
#used to extract date values from rows
pattern =r"([1-2][0-9][0-9][0-9])"
years=merged['SpecialNotes'].str.extract(pattern)
```

When we used the Series.str.extract() method, we enclosed our regular expression in parentheses. The parentheses indicate that only the character pattern matched should be extracted and returned in a series. We call this a capturing group.
<br>If the capturing group doesn't exist in a row (or there is no match) the value in that row is set to NaN instead. As a result, the Series returned looked like this:</br>
![ ](extract.png)

```python
merged = merged.set_index('Country')
```

```python
pattern = r"(?P<Years>[1-2][0-9]{3})"
merged['SpecialNotes'].str.extractall(pattern)
```

![ ](extractall.png)

Next, let's use the same regular expression from the last screen to extract all the years from the <b>Special Notes</b> column, except this time, we'll use a named capturing group. 
<br>Using a named capturing group means that we can refer to the group by the specified name instead of just a number. We can use the following syntax to add a name: </br>
<code> (?P<Column_Name>...) </code>

Below, we name the capturing group Years:

```python
#use of a named capturing group
pattern = r"(?P<Years>[1-2][0-9]{3})"
#Series.str.extractall() method to extract all of the years in the IESurvey
years = merged['IESurvey'].str.extractall(pattern)
#the Series.value_counts() method to create a list of the unique years, along with the count.
value_counts = years['Years'].value_counts()
```

```python
pattern = r"(?P<First_Year>[1-2][0-9]{3})/?(?P<Second_Year>[0-9]{2})?"

years = merged['IESurvey'].str.extractall(pattern)
first_two_year = years['First_Year'].str[0:2]
years['Second_Year'] = first_two_year + (years['Second_Year'])
```
